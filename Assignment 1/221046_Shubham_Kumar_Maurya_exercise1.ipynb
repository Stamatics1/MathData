#### Import NumPy as np

```python
import numpy as np
```

#### Create an array of 10 zeros 

```python
tenZero_array = np.zeros(10)
print(tenZero_array)
```

#### Create an array of 10 ones

```python
tenOnes_array = np.ones(10)
print(tenOnes_array)
```

#### Create an array of 10 fives

```python
tenFives_array = np.full(10,5)
print(tenFives_array)
```

#### Create an array of the integers from 10 to 50

```python
integers_array = np.arange(10,51)
print(integers_array)
```

#### Create an array of all the even integers from 10 to 50

```python
even_integers_array = np.arange(10,51,2)
print(even_integers_array)
```

#### Create a 3x3 matrix with values ranging from 0 to 8

```python
matrix = np.arange(9).reshape(3, 3)
print(matrix)
```

#### Create a 3x3 identity matrix

```python
identity_matrix = np.eye(3)
print(identity_matrix)
```

#### Use NumPy to generate a random number between 0 and 1

```python
a_random_number = np.random.rand()
print(a_random_number)
```

#### Use NumPy to generate an array of 25 random numbers sampled from a standard normal distribution

```python
array_of_25numbers = np.random.randn(25)
print(array_of_25numbers)
```

#### Create the following matrix:

```python
matrix = np.arange(0.01, 1.01, 0.01).reshape(10, 10)
print(matrix)
```

#### Create an array of 20 linearly spaced points between 0 and 1:

```python
linear_array = np.linspace(0, 1, 20)
print(linear_array)
```
