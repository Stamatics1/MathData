** Import pandas as pd.**

```python
import pandas as pd
```

** Read Salaries.csv as a dataframe called sal.**

```python
sal = pd.read_csv("Salaries.csv")
```

** Check the head of the DataFrame. **

```python
print(sal.head())
```

** Use the .info() method to find out how many entries there are.**

```python
print(sal.info())
```

**What is the average BasePay ?**

```python
average_base_pay = sal['BasePay'].mean()
print(average_base_pay)
```

** What is the highest amount of OvertimePay in the dataset ? **

```python
highest_overtime_pay = sal['OvertimePay'].max()
print(highest_overtime_pay)
```

** What is the job title of  JOSEPH DRISCOLL ? Note: Use all caps, otherwise you may get an answer that doesn't match up (there is also a lowercase Joseph Driscoll). **

```python
job_title = sal.loc[sal['EmployeeName'] == 'JOSEPH DRISCOLL', 'JobTitle'].values[0]
print(job_title)
```

** How much does JOSEPH DRISCOLL make (including benefits)? **

```python
total_pay = sal.loc[sal['EmployeeName'] == 'JOSEPH DRISCOLL', 'TotalPayBenefits'].values[0]
print(total_pay)
```

** What is the name of highest paid person (including benefits)?**

```python
highest_paid_index = sal['TotalPayBenefits'].idxmax()
highest_paid_person = sal.loc[highest_paid_index, 'EmployeeName']
print(highest_paid_person)
```

** What is the name of lowest paid person (including benefits)? Do you notice something strange about how much he or she is paid?**

```python
lowest_paid_index = sal['TotalPayBenefits'].idxmin()
lowest_paid_person = sal.loc[lowest_paid_index, 'EmployeeName']
print(lowest_paid_person)
```

** What was the average (mean) BasePay of all employees per year? (2011-2014) ? **

```python
sal['Year'] = pd.to_datetime(sal['Year'])
average_base_pay_per_year = sal.loc[(sal['Year'].dt.year >= 2011) & (sal['Year'].dt.year <= 2014)].groupby(sal['Year'].dt.year)['BasePay'].mean()
print(average_base_pay_per_year)
```

** How many unique job titles are there? **

```python
unique_job_titles = sal['JobTitle'].nunique()
print(unique_job_titles)
```

** What are the top 5 most common jobs? **

```python
top_5_common_jobs = sal['JobTitle'].value_counts().head(5)
print(top_5_common_jobs)
```

** How many Job Titles were represented by only one person in 2013? (e.g. Job Titles with only one occurence in 2013?) **

```python
job_titles_2013 = sal[sal['Year'] == 2013]['JobTitle'].value_counts()
job_titles_count = len(job_titles_2013[job_titles_2013 == 1])
print(job_titles_count)
```
