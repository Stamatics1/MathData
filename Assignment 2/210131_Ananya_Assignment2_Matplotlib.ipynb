```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
x = np.arange(0,100)
y = x*2
z = x**2

fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])  

ax.plot(x, y, 'b')
ax.set_xlabel('x') 
ax.set_ylabel('y')
ax.set_title('Title')
```

```python
 
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])  

 
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes1')
axes1.set_ylabel('Y_label_axes1')
axes1.set_title('Axes 1 Title')


axes2.plot(x, y, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title');
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1]) 
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.4]) 

 
 

```

```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])

axes1.plot(x, z, 'b')
axes1.axis('tight')
axes1.set_xlabel('X_label_axes1')
axes1.set_ylabel('Y_label_axes1')
axes1.set_title('Axes 1 Title')

axes2.plot(x, y, 'g')
axes2.set_xlabel('X_label_axes1')
axes2.set_ylabel('Y_label_axes1')
axes2.set_title('Axes 2 Title')
axes2.set_ylim([30, 50])
axes2.set_xlim([20, 22])
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, z, 'b--')
axes[0].set_xlabel('X_label_axes1')
axes[0].set_ylabel('Y_label_axes1')
axes[0].set_title('Axes 1 Title')

axes[1].plot(x, y, 'r')
axes[1].set_xlabel('X_label_axes1')
axes[1].set_ylabel('Y_label_axes1')
axes[1].set_title('Axes 2 Title')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15,3))

axes[0].plot(x, z, 'b--')
plt.tight_layout()
axes[0].set_xlabel('X_label_axes1')
axes[0].set_ylabel('Y_label_axes1')
axes[0].set_title('Axes 1 Title')

axes[1].plot(x, y, 'r')
axes[1].set_xlabel('X_label_axes1')
axes[1].set_ylabel('Y_label_axes1')
axes[1].set_title('Axes 2 Title')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()

sns.jointplot(x='fare',y='age',data=titanic,kind='scatter')
```

```python
 sns.histplot(titanic['fare'],kde=False,bins=30)
```

```python
sns.boxplot(x="class", y="age", data=titanic,palette='rainbow')
```

```python
sns.swarmplot(x="class", y="age", data=titanic, palette='rainbow')
```

```python
sns.countplot(x='sex',data=titanic)
```

```python
titanic.head()
titanic.corr()
sns.heatmap(titanic.corr())
```

```python
g = sns.FacetGrid(titanic, col="sex")
g = g.map(plt.hist, "age")
```
