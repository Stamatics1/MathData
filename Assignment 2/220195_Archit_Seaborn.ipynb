** Recreate the plots below using the titanic dataframe. There are very few hints since most of the plots can be done with just one or two lines of code and a hint would basically give away the solution. Keep careful attention to the x and y labels for hints.**

** *Note! In order to not lose the plot image, make sure you don't code in the cell that is directly above the plot, there is an extra cell above that one which won't overwrite that plot!* **

```python
g = sns.jointplot(x='fare', y='age', data=titanic)
g.plot_joint(sns.scatterplot)
plt.show()
```

```python
ax = sns.histplot(x='fare', data=titanic, bins=range(0, 901, 100), binwidth=100)

# Set the x-axis label
ax.set_xlabel('Fare')

# Set the x-axis ticks at the desired intervals
ax.set_xticks(range(0, 901, 100))

# Display the plot
plt.show()
```

```python
ax = sns.barplot(x='fare', y='age', data=titanic)
ax.set_xlabel('Fare')
plt.show()




```

```python
custom_palette = {'male': '#8cb5c5', 'female': '#a3c8a8'}

# Create the bar graph with 'sex' on the x-axis and count on the y-axis, using the custom colors
ax = sns.countplot(x='sex', data=titanic, palette=custom_palette)

# Set the y-axis ticks at the desired data points
ax.set_yticks([100, 200, 300, 400, 500, 600])

# Display the plot
plt.show()
```

```python
g = sns.FacetGrid(titanic, col='sex', height=5, aspect=1.5)
g.map_dataframe(sns.histplot, x='age', bins=range(0, 81, 10), color='navy')
g.set(xticks=[0, 10, 20, 30, 40, 50, 60, 70, 80], yticks=[0, 20, 40, 60, 80, 100, 120])

# Set the titles for each graph
g.set_titles(col_template='sex={col_name}')

# Display the plots
plt.show()

```
