## Exercise 3

** Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])
x_ticks = np.linspace(0, 1, 6)
y_ticks = np.linspace(0, 1, 6)
ax1.set_xticks(x_ticks)
ax1.set_yticks(y_ticks)
ax2.set_xticks(x_ticks)
ax2.set_yticks(y_ticks)
plt.show()
```

** Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
fig = plt.figure(figsize=(4.5, 7))
figsize=(20, 4)
ax1 = fig.add_axes([0,0,1,0.4])
ax2 = fig.add_axes([0.2,0.2,0.4,0.16])
ax3 = fig.add_axes([0,0.5, 1,0.4])
ax4 = fig.add_axes([0.2,0.7,0.4,0.16])

ax1.plot(x,x*x, 'blue')
ax2.plot(x,x*x, 'blue')
ax3.plot(x,x*x, 'blue')
ax4.plot(x,x*x, 'blue')

ax1.set_xlabel('X')
ax1.set_ylabel('Z')
ax2.set_xlabel('X')
ax2.set_ylabel('Z')
ax1.set_xlabel('X')
ax3.set_ylabel('Z')
ax4.set_xlabel('X')
ax4.set_ylabel('Z')

ax2.set_title('zoom')
ax4.set_title('zoom')
ax1.set_xlim(0,100)
ax3.set_xlim(0,100)
ax2.set_xlim(20,22)
ax2.set_ylim(300,500)
ax4.set_xlim(20,22)
ax4.set_ylim(300,500)
matplotlib.rcParams['font.size'] = 7
```

## Exercise 4

** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize = (4.5,3))
```

** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
# fig = plt.figure(figsize = (1,1))
fig, ax = plt.subplots(nrows=1, ncols=2, figsize = (4.5,3))
ax[0].plot(x,2*x, 'blue', linestyle = '--')
ax[1].plot(x,x*x , 'red')
for i in range(0,2):
  ax[i].set_xlabel('X')
  ax[i].set_xlabel('Y')
```
