## Exercise 1

** Follow along with these steps: **

## Exercise 3

** Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,0.4,0.4])
```

** Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
x1 = np.linspace(0, 100, 100)
y1 = x1 ** 2
x2 = np.linspace(20, 22, 100)
y2 = 2 * x2
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])
ax1.plot(x1, y1,'r')
ax1.set_xlabel('x')
ax1.set_ylabel('z')
ax1.set_ylim(bottom=0)
ax1.set_xlim(left=0)
ax2.plot(x2, y2,'r')
ax2.set_xlabel('x')
ax2.set_ylim(bottom=0)
ax2.set_xlim(left=20)
ax2.set_ylim(top=50)
ax2.set_xlim(right=22)
ax2.set_ylabel('y ')
```

## Exercise 4

** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
```

** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[1].set_ylim(bottom=0)
axes[1].set_xlim(left=0)
axes[1].set_ylim(top=10000)
axes[1].set_xlim(right=100)
axes[1].plot(x,x**2,'r--',lw=4)
axes[0].plot(x,x*2,'b',lw=6)
axes[0].set_ylim(bottom=0)
axes[0].set_xlim(left=0)
axes[0].set_ylim(top=200)
axes[0].set_xlim(right=100)

```

** See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
fig,axes=plt.subplots(figsize=(18,2),nrows=1,ncols=2)
axes[1].set_ylim(bottom=0)
axes[1].set_xlim(left=0)
axes[1].set_ylim(top=10000)
axes[1].set_xlim(right=100)
axes[1].plot(x,x**2,'r--',lw=4)
axes[0].plot(x,x*2,'b',lw=6)
axes[0].set_ylim(bottom=0)
axes[0].set_xlim(left=0)
axes[0].set_ylim(top=200)
axes[0].set_xlim(right=100)
```
