## Exercise 1

** Follow along with these steps: **
* ** Create a figure object called fig using plt.figure() **
* ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* ** Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

x = [1, 2, 3, 4, 5]

y = [2 * value for value in x]

ax.plot(x, y)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Plot of y = 2x')

plt.show()
```

** Now plot (x,y) on both axes. And call your figure object to show it.**

```python

# Create a figure object
fig = plt.figure()

# Add the first axis to the figure canvas
ax1 = fig.add_axes([0, 0, 1, 1])

# Add the second axis to the figure canvas
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])

# Define the x values
x = [1, 2, 3, 4, 5]

# Define the corresponding y values
y = [2 * value for value in x]

# Plot (x, y) on the first axis
ax1.plot(x, y)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('Plot on ax1')

# Plot (x, y) on the second axis
ax2.plot(x, y)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('Plot on ax2')

# Call the figure object to show the plot
fig
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])

ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])

x1 = np.linspace(0, 100, 1000)
y1 = np.exp(x1 * 0.1)
ax1.plot(x1, y1, color='blue')
ax1.set_xlim(0, 100)
ax1.set_ylim(0, 10000)
ax1.set_xlabel('X')
ax1.set_ylabel('Z')

x2 = np.linspace(20, 22, 1000)
y2 = np.exp(x2 * 0.1)
ax2.plot(x2, y2, color='red')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')

plt.show()
```
