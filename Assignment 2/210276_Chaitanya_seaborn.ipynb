```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!
sns.histplot(data=titanic, x='fare')
plt.ylim(0, 500)
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!
sns.boxplot(data=titanic, x='class', y='age')

# Show the plot
plt.show()
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!
sns.stripplot(data=titanic, x='class', y='age', jitter=-0.2)

# Show the plot
plt.show()
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!
sns.countplot(data=titanic, x='sex')

# Show the plot
plt.show()
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!
columns = ['survived', 'pclass', 'age', 'sibsp', 'parch', 'fare', 'adult_male', 'alone']

# Create the correlation matrix
corr_matrix = titanic[columns].corr()

# Create the colored square plot
sns.heatmap(corr_matrix, cmap='coolwarm')

# Show the plot
plt.show()


```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))

# Plot the histogram for males
sns.histplot(data=titanic[titanic['sex'] == 'male'], x='age', ax=ax1, color='blue')
ax1.set_title('sex=male')
ax1.set_xlabel('Age')
ax1.set_ylabel('Count')

# Plot the histogram for females
sns.histplot(data=titanic[titanic['sex'] == 'female'], x='age', ax=ax2, color='pink')
ax2.set_title('sex=female')
ax2.set_xlabel('Age')
ax2.set_ylabel('Count')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()








```
